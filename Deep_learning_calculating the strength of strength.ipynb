```python
#IMPORTING LIBRARIES
import pandas as pd
import numpy as np
from matplotlib import pyplot as pt
import sklearn
```

```python
#READING FILE
data=pd.read_csv("concrete_data.csv")
```

```python
# Arranging the table as per predictors and output
X=data.drop(columns=['Strength'])
Y=data[['Strength']]
X.head()
Y.head()
```

```python
# Modelling the Deep learning neuron structure
from keras.models import Sequential
from keras.layers import Dense
model=Sequential()
n_cols = X.shape[1]
model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
model.add(Dense(10, activation='relu'))
model.add(Dense(1))
```

```python
#Compiling the model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Splitting ,Training and Predicting the model and calculating the mean squared error
from sklearn.model_selection import train_test_split
list1=[]
for i in range(50):
   X_train,x_test,Y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
   model.fit(X_train,Y_train,epochs=50)
   test_y_predictions = model.predict(x_test)
   list1.append(sklearn.metrics.mean_squared_error(y_test, test_y_predictions,sample_weight=None, multioutput='uniform_average'))
print(list1) 
```

```python
#Calculating the mean and standard deviation
import statistics
print("The mean of mean squared errors",statistics.mean(list1))
print("The standard deviation of mean squared errors",statistics.stdev(list1))
```
